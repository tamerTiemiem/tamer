```python
# Finding distribution (nan if unknown)
ax = metadataCSV['finding'].value_counts(dropna=False).plot.pie(y='Finding', legend = True,
                autopct='%50.0f%%', figsize = (9.5,9.5), title = 'Finding Distribution')
```

```python
# Sex distribution (nan if unknown)
ax = metadataCSV['sex'].value_counts(dropna=False).plot.pie(y='sex', legend = True,
      autopct='%2.0f%%', figsize = (5,5), title = 'Sex Distribution')
```

```python
# Now age (nan in unknown)
out = pd.cut(metadataCSV['age'], bins=np.arange(0,110,10).tolist(), include_lowest=True)
ax = out.value_counts(sort=False, dropna=False).plot.bar(rot=0, color="b",
     figsize=(15,6), title= "Age Distribution")
#ax.set_xticklabels([c[1:-1].replace(","," to") for c in out.cat.categories])
plt.show()
```

```python
# Survival 
ax = metadataCSV['survival'].value_counts(dropna=False).plot.pie(y='survival', legend = True, autopct='%2.0f%%', figsize = (5,5), title = 'Survival Distribution')
```

```python
# View
ax = metadataCSV['view'].value_counts(dropna=False).plot.pie(y='view', legend = True, autopct='%2.0f%%', figsize = (8,8), title = 'View distribution')
```

```python
# Modality
ax = metadataCSV['modality'].value_counts(dropna=False).plot.pie(y='modality', legend = True, autopct='%2.0f%%', figsize = (5,5), title = 'Modality Distribution')
```

```python
# Location
metadataCSV['country'] = metadataCSV['location'].apply(lambda x: x.split(',')[-1].replace(" ","") if x is not np.nan else "Unknown")
ax = metadataCSV['country'].value_counts(dropna=False).plot.pie(y='country', legend = True, autopct='%2.0f%%', figsize = (10,10), title = 'Location of the patient')
```

```python
counter = 1
for (i , row) in metadataCSV.iterrows():
    if row["finding"] == "Pneumonia/Viral/COVID-19" and row["view"] == "PA" or row["view"] == "AP" :
        filename = row["filename"]
        img_path_old = os.path.join(path_dataSet_Pneumonia,filename)
        #print(img_path_old," : ",counter)
        img_path_New = os.path.join(target_folder_Pneumonia,filename)
        #print(img_path_New," : ",counter)
        shutil.copy2(img_path_old,img_path_New)
        counter+=1
print(" img copy ",counter-1)

        
```
